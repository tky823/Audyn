```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
from matplotlib.figure import Figure
from matplotlib.axes import Axes
```

```python
from audyn.functional import mobius_add, mobius_scalar_mul, mobius_sub
```

```python
plt.rcParams["font.size"] = 15
```

```python
def plot_arrow(fig: Figure, ax: Axes, arrow: torch.Tensor, text: str = "", linestyle: str = "-", color: str = "black") -> None:
    ax.annotate(
        "",
        xytext=(0, 0),
        xy=arrow.tolist(),
        arrowprops=dict(arrowstyle="->", linestyle=linestyle, color=color),
    )
    ax.annotate(
        text,
        xy=arrow.tolist(),
        color=color,
    )

def plot_circle(fig: Figure, ax: Axes, radius: float = 1) -> None:
    circle = patches.Circle((0, 0), radius, facecolor=(0, 0, 0, 0), edgecolor="gray")
    ax.add_patch(circle)
```

```python
x = torch.tensor([0.5, 0.0])
y = torch.tensor([-0.3, 0.4])
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

plot_circle(fig, ax)
plot_arrow(fig, ax, x, text=r"x", linestyle="-", color="red")
plot_arrow(fig, ax, y, text=r"y", linestyle="-", color="blue")

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

fig.show()
```

## Addition

$$
\mathbf{x} \oplus_{c} \mathbf{y}
$$

```python
x_plus_y = mobius_add(x, y)
y_plus_x = mobius_add(y, x)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

plot_circle(fig, ax)
plot_arrow(fig, ax, x, text=r"x", linestyle="-", color="red")
plot_arrow(fig, ax, y, text=r"y", linestyle="-", color="blue")
plot_arrow(fig, ax, x_plus_y, text=r"$x \oplus y$", linestyle="-", color="black")
plot_arrow(fig, ax, y_plus_x, text=r"$y \oplus x$", linestyle="--", color="black")

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

fig.show()
```

## Subtraction

$$
\mathbf{x} \ominus_{c} \mathbf{y}
$$

```python
x_minus_y = mobius_sub(x, y)
y_minus_x = mobius_sub(y, x)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

plot_circle(fig, ax)
plot_arrow(fig, ax, x, text=r"x", linestyle="-", color="red")
plot_arrow(fig, ax, y, text=r"y", linestyle="-", color="blue")
plot_arrow(fig, ax, x_minus_y, text=r"$x \ominus y$", linestyle="-", color="black")
plot_arrow(fig, ax, y_minus_x, text=r"$y \ominus x$", linestyle="--", color="black")

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

fig.show()
```

## Scaler multiplication

$$
\mathbf{x} \otimes_{c} r
$$

```python
x_mul_4 = mobius_scalar_mul(x, 4)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

plot_circle(fig, ax)
plot_arrow(fig, ax, x, text=r"x", linestyle="-", color="red")
plot_arrow(fig, ax, y, text=r"y", linestyle="-", color="blue")
plot_arrow(fig, ax, x_mul_4, text=r"$x \otimes 4$", linestyle="-", color="black")

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

fig.show()
```
